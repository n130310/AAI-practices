**Задача 1:**



Интернет-магазин предлагает следующие условия скидок:

*   Для заказов больше 1000 единиц, клиент получает скидку 5%. Если клиент использует промокод SUPERDISCOUNT, он получает скидку 10% (вместо 5%).
*  Для заказов более 5000 единиц, клиент получает скидку 15%, а использование промокода SUPERDISCOUNT увеличивает скидку до 20% (вместо 15%).

Этап 1:
Ввод:
```
Введите стоимость единицы товара: 5
Введите количество товара: 1001
Введите промокод: GiVEMEDISCONT
```

Вывод:

```
Ваша скидка: 5%
Итоговая сумма: 4754.75
```
Этап 2:

Оформите ваш код в виде функции

````python
price = int(input('Введите стоимость единицы товара: '));
bulk =  int(input('Введите количество товара: '));
promo = input('Введите промокод: ');

summa = 0;
sale = 'SUPERDISCOUNT';

if bulk > 1000 and bulk <= 5000 :
  summa = (bulk*price)*0.95;
elif bulk > 5000 :
  summa = (bulk*price)*0.85;
else :
  summa = (bulk*price);

if promo == sale :
  summa = summa*0.95;

sale = 5;

print('Ваша скидка: ', sale, '%');
print('Итоговая сумма: ', summa)
````

Оформите ваш код в виде функции

````python
numbers = list(input().split(' '));
newnumbers = [];
for i in numbers :
  newnumbers.append(int(i));

for i in range(len(newnumbers)) :
 if newnumbers[i] > 0 :
  numbers[i] = newnumbers [i];
 else :
  numbers[i] = None;

newnumbers = list(filter(None, numbers));
print(' '.join(map(str,newnumbers)))
````

**Задача 3:**

Этап 1:
Напишите программу реализующую Алгоритм Евклида


> Алгоритм Евклида – это алгоритм нахождения наибольшего общего делителя (НОД) пары целых чисел.

Чуть подробнее, если вы забыли: [link](https://skysmart.ru/articles/mathematic/naibolshij-obshchij-delitel)


Ввод:
```
30 18
```

Вывод:

```
6
```

Этап 2:
Оформите ваш код в виде функции


````python
i = []
for a in input().split():
    i.append( int(a) );

n1 = i[0];
n2 = i[1];
nod = 0;

while n2 != 0 :
    n1, n2 = n2, n1 % n2;
    nod = n1;

print(nod)
````

**Задача 4:**

Этап 1:
Напишите функцию программу, которая принимает строку и возвращает список слов и количество их упомнинаний в предложении

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
apple banana apple
```

Вывод:

```
apple: 2,
banana: 1
```

````python
words = list(input().split(' '));
wordlist = [];
wordcounter = [];
for word in words:
  if word not in wordlist:
    wordlist.append(word);
    wordcounter.append(0)

for i in range(len(words)):
  if word in wordlist:
    wordcounter[i] = wordcounter[i]+1;

print(wordcounter)
````
